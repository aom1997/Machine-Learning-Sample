## **Config & Functions**

```python
from datetime import date, timedelta
from pyspark.sql import functions as F
from pyspark.sql.window import Window
import pandas as pd

PROJECT     = 'your-gcp-project'
MSSP        = f'{PROJECT}.MSSP'
ML          = f'{PROJECT}.ML_DATASETS'
GCS_BUCKET  = 'your-gcp-project-temp'
CKPT_BASE = "dbfs:/tmp/checkpoints/dme_fraud_prediction"
ACO_ID      = 'your aco id' 

# Select Prediction or Training Dataset
Prediction = 1
if Prediction == 1:
    #start_date   = date(date.today().year-1,1,1) #claims start windows for prediction
    #end_date     = date.today() #claims start windows for prediction
    start_date   = date(2025,1,1)
    end_date   = date(2025,12,1)
    DESTINATION = f'{ML}.DME_Fraud_Features'
else:
    start_date   = date(2021,1,1) #claims start windows for training
    end_date     = date(2025,12,31)
    DESTINATION = f'{ML}.DME_Fraud_Training'

lookback_1yr = start_date - timedelta(days=365) #1year from most of features
lookback_2yr = start_date - timedelta(days=365*2) #2years for new bene features
lookback_5yr = start_date - timedelta(days=365*5) #5years for OrdeNPI Ever Seen Bene in Past 5 Years feature

def read_ml_table(full_table_name, filter_expr=None):
    #Read a BigQuery table directly into Spark — no credentials needed. 
    reader = spark.read.format('bigquery') \
        .option('table', full_table_name) \
        .option('parentProject', PROJECT)

    if filter_expr:
        reader = reader.option('filter', filter_expr)

    return reader.load()

def write_ml(df, full_table_name):
    #Write a Spark DataFrame to BigQuery via GCS temp bucket.
    df.write.format('bigquery') \
        .option('table', full_table_name) \
        .option('temporaryGcsBucket', GCS_BUCKET) \
        .mode('overwrite') \
        .save()
    print(f'Written → {full_table_name}')

print(f'start_date   : {start_date}')
print(f'end_date     : {end_date}')
print(f'lookback_1yr : {lookback_1yr}')
print(f'lookback_2yr : {lookback_2yr}')
print(f'lookback_5yr : {lookback_5yr}')
```

## **Step 1 — Build dme_pred & dme_header_pred**

```python
# ── Read assigned beneficiaries ───────────────────────────────
df_assign = read_ml_table(
    f'{MSSP}.CURRENTASSIGNMENTSMATERIALIZED',
    filter_expr=f"ACO_UNIQUE_ID = '{ACO_ID}' AND ASSIGNMENTYEAR >= '{start_date.year}'"
).select('unique_id').distinct()

# ── Build dme_line_2yr (filter at read time, no full table scan) ──
# ── Use Assigned Patients only for Training, All Patients for Prediction ──
df_dme_line_2yr = read_ml_table(
    f'{MSSP}.PartBDME_Line_Items',
    filter_expr=f"clm_thru_dt >= '{lookback_2yr}' and clm_thru_dt <= '{end_date}' and aco_unique_id = '{ACO_ID}'"
).dropDuplicates(['CUR_CLM_UNIQ_ID', 'CLM_LINE_NUM'])



# ── Build dme_header_pred (one row per claim look back 2 years) ─────────────────
df_dme_header_2yr= df_dme_line_2yr \
    .groupBy('CUR_CLM_UNIQ_ID') \
    .agg(
        F.max('CLM_FROM_DT').alias('CLM_FROM_DT'),
        F.max('clm_thru_dt').alias('clm_thru_dt'),
        F.max('ORDRG_PRVDR_NPI_NUM').alias('ordrg_prvdr_npi_num'),
        F.max('PAYTO_PRVDR_NPI_NUM').alias('pay_to_prvdr_npi_num'),
        F.max('unique_id').alias('Beneficiary_ID')
    )

# ── Build dme_line_1yr (one row per claim look back 1 years) ─────────────────
df_dme_line_1yr = df_dme_line_2yr.filter(
    (F.col('clm_thru_dt') >= lookback_1yr) & (F.col('clm_thru_dt') <= end_date)
)

df_dme_line_1yr.cache()

# ── Build dme_header_1yr (one row per claim look back 1 years) ─────────────────
df_dme_header_1yr= df_dme_line_1yr \
    .groupBy('CUR_CLM_UNIQ_ID') \
    .agg(
        F.max('CLM_FROM_DT').alias('CLM_FROM_DT'),
        F.max('clm_thru_dt').alias('clm_thru_dt'),
        F.max('ORDRG_PRVDR_NPI_NUM').alias('ordrg_prvdr_npi_num'),
        F.max('PAYTO_PRVDR_NPI_NUM').alias('pay_to_prvdr_npi_num'),
        F.max('unique_id').alias('Beneficiary_ID')
    )

# ── Build dme_index (only include claims within start and end date) ─────────────────
df_index = df_dme_header_1yr.filter(
    (F.col('clm_thru_dt') >= start_date) & (F.col('clm_thru_dt') <= end_date)
    )

df_index.cache()
```

## **Step 2 — Build Claim Level Feature Table**


```python
# ── Read high risk CPT codes ──────────────────────────────────
df_high_risk_cpt = read_ml_table(f'{ML}.DME_Fraud_RiskCPT').select('CPT').distinct()
high_risk_set = set([row['CPT'] for row in df_high_risk_cpt.collect()])
print(f'High risk CPTs: {len(high_risk_set)}')

# ── Filter to prediction window ───────────────────────────────
df_features_claim = df_dme_line_1yr.filter(
    (F.col('clm_thru_dt') >= start_date) &
    (F.col('clm_thru_dt') <= end_date)
)

# ── Aggregate to claim header level ──────────────────────────
df_features_claim = df_features_claim \
    .groupBy('cur_clm_uniq_id') \
    .agg(
        F.sum('CLM_LINE_ALLOWD_CHRG_AMT').alias('clm_chrg_amt'),
        F.sum('CLM_LINE_CVRD_PD_AMT').alias('clm_paid_amt'),
        F.count('clm_line_num').alias('clm_lines_ct'),
        F.collect_set('clm_line_hcpcs_cd').alias('clm_all_cpts'),
        F.countDistinct('clm_line_hcpcs_cd').alias('clm_uniq_cpt_ct'),
        F.min('clm_from_dt').alias('clm_from_dt'),
        F.max('clm_thru_dt').alias('clm_thru_dt'),
        F.max('clm_adjsmt_type_cd').alias('clm_status'),
        F.max('PAYTO_PRVDR_NPI_NUM').alias('clm_billnpi'),
        F.max('ordrg_prvdr_npi_num').alias('clm_ordenpi'),
        F.max('Unique_ID').alias('clm_beneficiary_id'),
    )
    
# ── Date features ─────────────────────────────────────────────
df_features_claim = df_features_claim \
    .withColumn('clm_from_month',    F.month('clm_from_dt')) \
    .withColumn('clm_from_day',      F.dayofweek('clm_from_dt')) \
    .withColumn('clm_duration_days', F.datediff('clm_thru_dt', 'clm_from_dt') + 1) \
    .withColumn('original_clm_flag', F.when(F.col('clm_status') == 0, 0).otherwise(1)) \
    .drop('clm_status')

# ── High risk CPT flag ────────────────────────────────────────
high_risk_array = F.array(*[F.lit(cpt) for cpt in high_risk_set])

df_features_claim = df_features_claim \
    .withColumn('risk_cpt_flag',
        F.when(F.size(F.array_intersect(F.col('clm_all_cpts'), high_risk_array)) > 0, 1)
        .otherwise(0)
    )

df_features_claim = df_features_claim \
    .withColumn('clm_all_cpts', F.concat_ws(',', F.col('clm_all_cpts')))

CKPT_FEATURES_CLAIM = f"{CKPT_BASE}/features_claim"
df_features_claim.write.mode("overwrite").parquet(CKPT_FEATURES_CLAIM)
df_features_claim = spark.read.parquet(CKPT_FEATURES_CLAIM)

print(f'Claim features: {df_features_claim.count():,} rows')
```

## **Step 3 — Build Beneficiary Level Feature Table**

```python
# ── Join for rolling lookback ─────────────────────────────────
df_bene_join = df_index.alias('a') \
    .join(
        df_dme_line_1yr.hint("range_join", 10).alias('b'),
        on=(
            (F.col('a.Beneficiary_ID') == F.col('b.Unique_ID')) &
            (F.col('b.clm_thru_dt') <= F.col('a.clm_thru_dt')) &
            (F.col('b.clm_thru_dt') >= F.date_sub(F.col('a.clm_thru_dt'), 365))
        ),
        how='inner'
    ) \
    .join(
        df_high_risk_cpt.alias('c'),
        on=(F.col('b.clm_line_hcpcs_cd') == F.col('c.CPT')),
        how='left'
    ) \
    .select(
        F.col('a.CUR_CLM_UNIQ_ID'),
        F.col('a.clm_thru_dt').alias('a_clm_thru_dt'),
        F.col('a.pay_to_prvdr_npi_num'),
        F.col('a.ordrg_prvdr_npi_num'),
        F.col('a.Beneficiary_ID'),
        F.col('b.clm_thru_dt').alias('b_clm_thru_dt'),
        F.col('b.CUR_CLM_UNIQ_ID').alias('b_CUR_CLM_UNIQ_ID'),
        F.col('b.clm_line_hcpcs_cd'),
        F.col('b.CLM_LINE_ALLOWD_CHRG_AMT'),
        F.col('b.CLM_LINE_CVRD_PD_AMT'),
        F.col('b.ordrg_prvdr_npi_num').alias('b_ordrg_prvdr_npi_num'),
        F.col('b.PAYTO_PRVDR_NPI_NUM'),
        F.col('c.CPT'),
    )


# ── BENE_FEATURES_1 ───────────────────────────────────────────
df_bene_f1 = df_bene_join \
    .groupBy('CUR_CLM_UNIQ_ID') \
    .agg(
        F.count(F.when(F.col('b_clm_thru_dt') >= F.date_sub(F.col('a_clm_thru_dt'), 30),  F.col('b_CUR_CLM_UNIQ_ID'))).alias('bene_clmline_ct_30days'),
        F.count(F.when(F.col('b_clm_thru_dt') >= F.date_sub(F.col('a_clm_thru_dt'), 90),  F.col('b_CUR_CLM_UNIQ_ID'))).alias('bene_clmline_ct_90days'),
        F.count(F.when(F.col('b_clm_thru_dt') >= F.date_sub(F.col('a_clm_thru_dt'), 365), F.col('b_CUR_CLM_UNIQ_ID'))).alias('bene_clmline_ct_365days'),
        F.countDistinct(F.when(F.col('b_clm_thru_dt') >= F.date_sub(F.col('a_clm_thru_dt'), 30),  F.col('b_CUR_CLM_UNIQ_ID'))).alias('bene_clmheader_ct_30days'),
        F.countDistinct(F.when(F.col('b_clm_thru_dt') >= F.date_sub(F.col('a_clm_thru_dt'), 90),  F.col('b_CUR_CLM_UNIQ_ID'))).alias('bene_clmheader_ct_90days'),
        F.countDistinct(F.when(F.col('b_clm_thru_dt') >= F.date_sub(F.col('a_clm_thru_dt'), 365), F.col('b_CUR_CLM_UNIQ_ID'))).alias('bene_clmheader_ct_365days'),
        F.countDistinct(F.when(F.col('b_clm_thru_dt') >= F.date_sub(F.col('a_clm_thru_dt'), 30),  F.col('clm_line_hcpcs_cd'))).alias('bene_uniq_cpt_ct_30days'),
        F.countDistinct(F.when(F.col('b_clm_thru_dt') >= F.date_sub(F.col('a_clm_thru_dt'), 90),  F.col('clm_line_hcpcs_cd'))).alias('bene_uniq_cpt_ct_90days'),
        F.countDistinct(F.when(F.col('b_clm_thru_dt') >= F.date_sub(F.col('a_clm_thru_dt'), 365), F.col('clm_line_hcpcs_cd'))).alias('bene_uniq_cpt_ct_365days'),
        F.sum(F.when(F.col('b_clm_thru_dt') >= F.date_sub(F.col('a_clm_thru_dt'), 30),  F.col('CLM_LINE_ALLOWD_CHRG_AMT'))).alias('bene_chrg_amt_30days'),
        F.sum(F.when(F.col('b_clm_thru_dt') >= F.date_sub(F.col('a_clm_thru_dt'), 90),  F.col('CLM_LINE_ALLOWD_CHRG_AMT'))).alias('bene_chrg_amt_90days'),
        F.sum(F.when(F.col('b_clm_thru_dt') >= F.date_sub(F.col('a_clm_thru_dt'), 365), F.col('CLM_LINE_ALLOWD_CHRG_AMT'))).alias('bene_chrg_amt_365days'),
        F.sum(F.when(F.col('b_clm_thru_dt') >= F.date_sub(F.col('a_clm_thru_dt'), 30),  F.col('CLM_LINE_CVRD_PD_AMT'))).alias('bene_paid_amt_30days'),
        F.sum(F.when(F.col('b_clm_thru_dt') >= F.date_sub(F.col('a_clm_thru_dt'), 90),  F.col('CLM_LINE_CVRD_PD_AMT'))).alias('bene_paid_amt_90days'),
        F.sum(F.when(F.col('b_clm_thru_dt') >= F.date_sub(F.col('a_clm_thru_dt'), 365), F.col('CLM_LINE_CVRD_PD_AMT'))).alias('bene_paid_amt_365days'),
        F.sum(F.when((F.col('b_clm_thru_dt') >= F.date_sub(F.col('a_clm_thru_dt'), 30))  & F.col('CPT').isNotNull(), F.lit(1)).otherwise(0)).alias('bene_risk_cpt_ct_30days'),
        F.sum(F.when((F.col('b_clm_thru_dt') >= F.date_sub(F.col('a_clm_thru_dt'), 90))  & F.col('CPT').isNotNull(), F.lit(1)).otherwise(0)).alias('bene_risk_cpt_ct_90days'),
        F.sum(F.when((F.col('b_clm_thru_dt') >= F.date_sub(F.col('a_clm_thru_dt'), 365)) & F.col('CPT').isNotNull(), F.lit(1)).otherwise(0)).alias('bene_risk_cpt_ct_365days'),
        F.countDistinct(F.when(F.col('b_clm_thru_dt') >= F.date_sub(F.col('a_clm_thru_dt'), 30),  F.col('b_ordrg_prvdr_npi_num'))).alias('bene_uniq_ordenpi_30days'),
        F.countDistinct(F.when(F.col('b_clm_thru_dt') >= F.date_sub(F.col('a_clm_thru_dt'), 90),  F.col('b_ordrg_prvdr_npi_num'))).alias('bene_uniq_ordenpi_90days'),
        F.countDistinct(F.when(F.col('b_clm_thru_dt') >= F.date_sub(F.col('a_clm_thru_dt'), 365), F.col('b_ordrg_prvdr_npi_num'))).alias('bene_uniq_ordenpi_365days'),
        F.countDistinct(F.when(F.col('b_clm_thru_dt') >= F.date_sub(F.col('a_clm_thru_dt'), 30),  F.col('PAYTO_PRVDR_NPI_NUM'))).alias('bene_uniq_billnpi_30days'),
        F.countDistinct(F.when(F.col('b_clm_thru_dt') >= F.date_sub(F.col('a_clm_thru_dt'), 90),  F.col('PAYTO_PRVDR_NPI_NUM'))).alias('bene_uniq_billnpi_90days'),
        F.countDistinct(F.when(F.col('b_clm_thru_dt') >= F.date_sub(F.col('a_clm_thru_dt'), 365), F.col('PAYTO_PRVDR_NPI_NUM'))).alias('bene_uniq_billnpi_365days'),
    )

# ── BENE_FEATURES_2 ───────────────────────────────────────────
df_bene_f2 = df_index.alias('a') \
    .join(
        df_dme_line_1yr.hint("range_join", 10).alias('b'),
        on=(
            (F.col('a.Beneficiary_ID') == F.col('b.Unique_ID')) &
            (F.col('b.clm_thru_dt') <= F.col('a.clm_thru_dt')) &
            (F.col('b.clm_thru_dt') >= F.date_sub(F.col('a.clm_thru_dt'), 365)) &
            (F.col('b.CUR_CLM_UNIQ_ID') != F.col('a.CUR_CLM_UNIQ_ID'))
        ),
        how='left'
    ) \
    .groupBy('a.CUR_CLM_UNIQ_ID') \
    .agg(
        F.min(F.datediff(F.col('a.clm_thru_dt'), F.col('b.clm_thru_dt'))).alias('bene_days_last_claim')
    )

# ── BENE_FEATURES_3 ───────────────────────────────────────────
df_bene_f3 = df_index.alias('a') \
    .join(
        df_dme_line_1yr.hint("range_join", 10).alias('b'),
        on=(
            (F.col('a.Beneficiary_ID') == F.col('b.Unique_ID')) &
            (F.col('b.clm_thru_dt') < F.col('a.clm_thru_dt')) &
            (F.col('b.clm_thru_dt') >= F.date_sub(F.col('a.clm_thru_dt'), 365))
        ),
        how='left'
    ) \
    .groupBy('a.CUR_CLM_UNIQ_ID') \
    .agg(
        F.countDistinct(F.when((F.col('b.clm_thru_dt') >= F.date_sub(F.col('a.clm_thru_dt'), 30))  & (F.col('a.pay_to_prvdr_npi_num') == F.col('b.PAYTO_PRVDR_NPI_NUM')), F.col('b.CUR_CLM_UNIQ_ID'))).alias('bene_seen_billnpi_ct_30days'),
        F.countDistinct(F.when((F.col('b.clm_thru_dt') >= F.date_sub(F.col('a.clm_thru_dt'), 90))  & (F.col('a.pay_to_prvdr_npi_num') == F.col('b.PAYTO_PRVDR_NPI_NUM')), F.col('b.CUR_CLM_UNIQ_ID'))).alias('bene_seen_billnpi_ct_90days'),
        F.countDistinct(F.when((F.col('b.clm_thru_dt') >= F.date_sub(F.col('a.clm_thru_dt'), 365)) & (F.col('a.pay_to_prvdr_npi_num') == F.col('b.PAYTO_PRVDR_NPI_NUM')), F.col('b.CUR_CLM_UNIQ_ID'))).alias('bene_seen_billnpi_ct_365days'),
        F.countDistinct(F.when((F.col('b.clm_thru_dt') >= F.date_sub(F.col('a.clm_thru_dt'), 30))  & (F.col('a.ordrg_prvdr_npi_num') == F.col('b.ordrg_prvdr_npi_num')), F.col('b.CUR_CLM_UNIQ_ID'))).alias('bene_seen_ordenpi_ct_30days'),
        F.countDistinct(F.when((F.col('b.clm_thru_dt') >= F.date_sub(F.col('a.clm_thru_dt'), 90))  & (F.col('a.ordrg_prvdr_npi_num') == F.col('b.ordrg_prvdr_npi_num')), F.col('b.CUR_CLM_UNIQ_ID'))).alias('bene_seen_ordenpi_ct_90days'),
        F.countDistinct(F.when((F.col('b.clm_thru_dt') >= F.date_sub(F.col('a.clm_thru_dt'), 365)) & (F.col('a.ordrg_prvdr_npi_num') == F.col('b.ordrg_prvdr_npi_num')), F.col('b.CUR_CLM_UNIQ_ID'))).alias('bene_seen_ordenpi_ct_365days'),
    )

# ── BENE_FEATURES_4 ───────────────────────────────────────────
parta_base = read_ml_table(f'{MSSP}.PartA_Claims_Adjusted_materialized') \
    .filter((F.col('clm_thru_dt') >= str(lookback_5yr)) & (F.col('clm_thru_dt') <= str(end_date)))

df_parta = parta_base \
    .select(F.col('Unique_ID'), F.col('OPRTG_PRVDR_NPI_NUM').alias('Provider'), F.col('clm_thru_dt')) \
    .union(
        parta_base.select(F.col('Unique_ID'), F.col('ATNDG_PRVDR_NPI_NUM').alias('Provider'), F.col('clm_thru_dt'))
    ) \
    .distinct()

df_partb = read_ml_table(f'{MSSP}.PartB_Materialized') \
    .filter((F.col('clm_thru_dt') >= str(lookback_5yr)) & (F.col('clm_thru_dt') <= str(end_date))) \
    .select(F.col('Unique_ID'), F.col('RNDRG_PRVDR_NPI_NUM').alias('Provider'), F.col('clm_thru_dt')) \
    .distinct()

df_seen = df_parta.union(df_partb).distinct()

df_bene_f4_seen = df_index.alias('a') \
    .join(
        df_seen.hint("range_join", 30).alias('b'),
        on=(
            (F.col('a.Beneficiary_ID') == F.col('b.Unique_ID')) &
            (F.col('a.ordrg_prvdr_npi_num') == F.col('b.Provider')) &
            (F.col('b.clm_thru_dt') >= F.add_months(F.col('a.clm_thru_dt'), -60)) &
            (F.col('b.clm_thru_dt') <= F.col('a.clm_thru_dt'))
        ),
        how='inner'
    ) \
    .select('a.CUR_CLM_UNIQ_ID').distinct() \
    .withColumn('bene_seen_ordenpi_5years', F.lit(0))

df_bene_f4 = df_index.select('CUR_CLM_UNIQ_ID') \
    .join(df_bene_f4_seen, on='CUR_CLM_UNIQ_ID', how='left') \
    .withColumn('bene_seen_ordenpi_5years', F.when(F.col('bene_seen_ordenpi_5years').isNull(), F.lit(0)).otherwise(F.lit(1)))

# ── Merge all bene features ───────────────────────────────────
df_features_bene = df_bene_f1 \
    .join(df_bene_f2, on='CUR_CLM_UNIQ_ID', how='left') \
    .join(df_bene_f3, on='CUR_CLM_UNIQ_ID', how='left') \
    .join(df_bene_f4, on='CUR_CLM_UNIQ_ID', how='left')

# ── Calculated fields ─────────────────────────────────────────
for w in [30, 90, 365]:
    df_features_bene = df_features_bene \
        .withColumn(f'bene_avg_chrg_amt_per_claim_{w}days', F.col(f'bene_chrg_amt_{w}days') / F.col(f'bene_clmheader_ct_{w}days')) \
        .withColumn(f'bene_avg_paid_amt_per_claim_{w}days', F.col(f'bene_paid_amt_{w}days')  / F.col(f'bene_clmheader_ct_{w}days'))

df_features_bene = df_features_bene \
    .withColumn('bene_bill_ord_pct_30days',  F.col('bene_uniq_billnpi_30days')  / F.col('bene_uniq_ordenpi_30days')) \
    .withColumn('bene_bill_ord_pct_90days',  F.col('bene_uniq_billnpi_90days')  / F.col('bene_uniq_ordenpi_90days')) \
    .withColumn('bene_bill_ord_pct_365days', F.col('bene_uniq_billnpi_365days') / F.col('bene_uniq_ordenpi_365days'))

CKPT_FEATURES_BENE = f"{CKPT_BASE}/features_bene"
df_features_bene.write.mode("overwrite").parquet(CKPT_FEATURES_BENE)
df_features_bene = spark.read.parquet(CKPT_FEATURES_BENE)
print(f'Bene features: {df_features_bene.count():,} rows')
```

## **Step 7 — Build Final Feature Table**

```python
# ── Merge all feature tables ──────────────────────────────────
df_tran = df_features_claim \
    .join(df_features_bene,        on='CUR_CLM_UNIQ_ID', how='left') \
    .join(df_features_billnpi,     on='CUR_CLM_UNIQ_ID', how='left') \
    .join(df_features_billnpibene, on='CUR_CLM_UNIQ_ID', how='left') \
    .join(df_features_ordenpi,      on='CUR_CLM_UNIQ_ID', how='left') \
    .join(df_features_ordenpibene,  on='CUR_CLM_UNIQ_ID', how='left')

# ── Flag Fraud Claims for Training Dataset, Filter to Assigned Patients only for Training Dataset ──────────────────────────────────
if Prediction == 1:
    df_tran = df_tran \
    # df_tran = df_tran.alias('a') \
    #     .join(
    #         df_assign.alias('b'),
    #         on='unique_id',
    #         how='inner'
    #     ) \
    #     .select(F.col('a.*')) \
    #     .withColumn('fraud_flag', F.lit(None).cast('integer'))  # or a sensible default
else:
    df_fraud_companies = read_ml_table('your-gcp-project.ML_DATASETS.DME_Fraudulent_Flags') \
        .filter(F.col('Status') == 'Submitted')

    df_tran = df_tran.alias('a') \
        .join(
            df_fraud_companies.alias('b'),
            on=F.col('a.clm_billnpi') == F.col('b.NPI'),
            how='left'
        ) \
        .withColumn('fraud_flag',
            F.when(F.col('b.NPI').isNull(), 0).otherwise(1)) \
        .select(F.col('a.*'), F.col('fraud_flag'))  # keep fraud_flag after dropping b.*

# ── Final calculated fields ───────────────────────────────────
df_tran = df_tran \
    .withColumn('billnpibene_pct_claimline_on_this_patient_365days',
                F.col('billnpibene_clmline_ct_365days')   / F.nullif(F.col('billnpi_clmline_ct_365days'),   F.lit(0))) \
    .withColumn('billnpibene_pct_claimheader_on_this_patient_365days',
                F.col('billnpibene_clmheader_ct_365days') / F.nullif(F.col('billnpi_clmheader_ct_365days'), F.lit(0))) \
    .withColumn('billnpibene_pct_paid_on_this_patient_365days',
                F.col('billnpibene_paid_amt_365days')     / F.nullif(F.col('billnpi_paid_amt_365days'),     F.lit(0))) \
    .withColumn('billnpibene_pct_chrg_on_this_patient_365days',
                F.col('billnpibene_chrg_amt_365days')     / F.nullif(F.col('billnpi_chrg_amt_365days'),     F.lit(0))) \
    .withColumn('ordenpibene_pct_claimline_on_this_patient_365days',
                F.col('ordenpibene_clmline_ct_365days')   / F.nullif(F.col('ordenpi_clmline_ct_365days'),   F.lit(0))) \
    .withColumn('ordenpibene_pct_claimheader_on_this_patient_365days',
                F.col('ordenpibene_clmheader_ct_365days') / F.nullif(F.col('ordenpi_clmheader_ct_365days'), F.lit(0))) \
    .withColumn('ordenpibene_pct_paid_on_this_patient_365days',
                F.col('ordenpibene_paid_amt_365days')     / F.nullif(F.col('ordenpi_paid_amt_365days'),     F.lit(0))) \
    .withColumn('ordenpibene_pct_chrg_on_this_patient_365days',
                F.col('ordenpibene_chrg_amt_365days')     / F.nullif(F.col('ordenpi_chrg_amt_365days'),     F.lit(0))) \
    .withColumn('billnpi_interaction_90days_x_top5cpt',
                F.col('billnpi_90days_claim_percentage_to_365days') * F.col('billnpi_top5_cpt_percentage_365days')) \
    .withColumn('ordenpi_interaction_90days_x_top5cpt',
                F.col('ordenpi_90days_claim_percentage_to_365days') * F.col('ordenpi_top5_cpt_percentage_365days')) \
    .withColumn('dataset', F.lit(date.today().strftime('%Y-%m-%d')))

# ── Write final output to BigQuery ────────────────────────────
# cache() means write reads from memory, not re-executing joins
write_ml(df_tran, DESTINATION)
print(f'Done → {DESTINATION}')
```
